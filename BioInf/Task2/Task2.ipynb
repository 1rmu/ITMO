### Implement RandomizedMotifSearch

```python
import random

def iterate_rand_search(dna, k ,t):
    i = 0
    best_motifs = []
    for i in range(t):
        best_motifs.append(dna[i][:k])
    while i < 1000:
        motifs = rand_search(dna, k, t)
        if score(motifs) < score(best_motifs):
            best_motifs = motifs
        i = i+1
    return best_motifs


def rand_search(dna, k, t):
    best_motifs = rand_motifs(dna, k)
    while True:
        prof = create_prof(best_motifs)
        motifs = create_motifs(prof, dna)
        if score(motifs) < score(best_motifs):
            best_motifs = motifs
        else:
            return best_motifs


def create_motifs(prof, dna):
    motifs = []
    for i in dna:
        motifs.append(prof_most_probable(i, prof, len(prof)))
    return motifs


def calc_probablity(prof, str):
    probablity = 1
    for i in range(0, len(str)):
        if str[i] == 'A':
            probablity = probablity * prof[i][0]
        elif str[i] == 'C':
            probablity = probablity * prof[i][1]
        elif str[i] == 'G':
            probablity = probablity * prof[i][2]
        elif str[i] == 'T':
            probablity = probablity * prof[i][3]
    return probablity


def prof_most_probable(dna_str, prof, k):
    best_pattern = dna_str[0:0 + k]
    best_probability = 0
    for i in range(len(dna_str) - k + 1):
        str = dna_str[i:i + k]
        new_probablity = calc_probablity(prof, str)
        if new_probablity > best_probability:
            best_pattern = str
            best_probability = new_probablity
    return best_pattern


def create_prof(motifs):
    profile =[]
    for i in range(len(motifs[0])):
        count_A, count_C, count_G, count_T = 1, 1, 1, 1
        for motif in motifs:
            if motif[i] == 'A':
                count_A += 1
            elif motif[i] == 'C':
                count_C += 1
            elif motif[i] == 'G':
                count_G += 1
            elif motif[i] == 'T':
                count_T += 1
        profile.append([count_A / (len(motifs) + 4), count_C/ (len(motifs) + 4),
                        count_G / (len(motifs) + 4), count_T / (len(motifs) + 4)])
    return profile


def ham_dist(str1, str2):
    coun = 0
    for s1, s2 in zip(str1, str2):
        if s1 != s2:
            coun += 1
    return coun


def f_consensus(motifs):
    consensus = ''
    for i in range(len(motifs[0])):
        count_A, count_C, count_G, count_T = 0, 0, 0, 0
        for motif in motifs:
            if motif[i] == 'A':
                count_A += 1
            elif motif[i] == 'C':
                count_C += 1
            elif motif[i] == 'G':
                count_G += 1
            elif motif[i] == 'T':
                count_T += 1
        if count_A >= max(count_C, count_G, count_T):
            consensus += "A"
        elif count_C >= max(count_A, count_G, count_T):
            consensus += "C"
        elif count_G >= max(count_C, count_A, count_T):
            consensus += "G"
        elif count_T >= max(count_C, count_G, count_A):
            consensus += "T"
    return consensus


def score(motifs):
    consensus = f_consensus(motifs)
    score = 0
    for motif in motifs:
        score += ham_dist(consensus, motif)
    return score


def rand_motifs(dna,k):
    motifs = []
    for i in dna:
        point = random.randint(0, len(i)-k)
        motifs.append(i[point:point + k])
    return motifs

if __name__ == "__main__":
    data = "".join(open('rosalind_ba2f.txt')).split()
    best_motifs = iterate_rand_search(data[2:], int(data[0]), int(data[1]))
    for i in best_motifs:
        print(i)
```
